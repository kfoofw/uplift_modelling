```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./criteo_data/criteo-uplift-v2.1.csv")
```

```python
df['treatment'].value_counts(normalize=True)
```

```python
# 1 percent of data
X_data = df.drop(columns = ["conversion"])
y_data = df[["conversion"]]
_, X_sample, _, y_sample = train_test_split(X_data, y_data, test_size = 0.05, random_state = 0, stratify = X_data[["treatment"]])
```

```python
X_sample
```

```python
y_sample
```

```python
sampled_df = pd.merge(X_sample, y_sample, left_index = True, right_index = True)
```

```python
sampled_df
```

```python
sampled_df['treatment'].value_counts(normalize=True)
```

```python
# Output data
sampled_df.to_csv("./criteo_data/sampled_df.csv")
```
